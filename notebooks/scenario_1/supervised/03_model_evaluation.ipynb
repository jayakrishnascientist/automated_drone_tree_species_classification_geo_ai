# 03 Model Evaluation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
```

## Load predictions and ground truth

```python
PRED_CSV = "outputs/predictions.csv"
GT_CSV = "data/labels/ground_truth.csv"

df_pred = pd.read_csv(PRED_CSV)
df_true = pd.read_csv(GT_CSV)

df = df_pred.merge(df_true, on="image_name")
```

## Confusion matrix

```python
label_map = {"acacia":0,"non_acacia":1}

y_true = df["label_y"].map(label_map)
y_pred = df["label_x"].map(label_map)

cm = confusion_matrix(y_true,y_pred)

sns.heatmap(cm,annot=True,fmt="d",
            xticklabels=["acacia","non_acacia"],
            yticklabels=["acacia","non_acacia"])

plt.title("Confusion Matrix")
plt.show()
```
